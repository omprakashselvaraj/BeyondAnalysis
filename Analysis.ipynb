```python
def cat1(x):
    dic={'M':1 ,'I':2, 'C':3, 'B':4, 'K':5, 'G':6, 'A':7, 'J':8, 'L':9, 'F':10, 'E':11, 'D':12}
    return dic[x]
```

```python
def cat2(x):
    dic={'D':1, 'C':2, 'F':3, 'I':4, 'B':5, 'A':6, 'E':7, 'H':8, 'G':9}
    return dic[x]
```

```python
train['CATEGORY_1']=train['CATEGORY_1'].apply(cat1)
test['CATEGORY_1']=test['CATEGORY_1'].apply(cat1)
```

```python
train['CATEGORY_2']=train['CATEGORY_2'].apply(cat2)
test['CATEGORY_2']=test['CATEGORY_2'].apply(cat2)
```
